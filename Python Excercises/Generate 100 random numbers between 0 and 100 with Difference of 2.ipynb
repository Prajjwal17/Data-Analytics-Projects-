```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 random numbers between 0 and 100
x = np.random.randint(0, 100, 100)

# Create a new array y, where each element is the previous element plus 2
y = np.zeros(len(x))
y[0] = x[0]
for i in range(1, len(x)):
  y[i] = y[i - 1] + 2

# Plot the two arrays
plt.plot(x, y)

# Set the title and labels
plt.title('Random Variables with Difference of 2')
plt.xlabel('X')
plt.ylabel('Y')

# Show the plot
plt.show()
```
